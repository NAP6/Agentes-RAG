```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Supongamos que estos son los embeddings ya obtenidos (vectores de ejemplo aleatorios)
```

```python
embeddings_dic = {
    'Renault': np.array(embeddings.get_text_embedding('Renault')),
    'Volkswagen': np.array(embeddings.get_text_embedding('Volkswagen')),
    'BMW': np.array(embeddings.get_text_embedding('BMW')),
    'Mercedes-Benz': np.array(embeddings.get_text_embedding('Mercedes-Benz')),
    'Toyota': np.array(embeddings.get_text_embedding('Toyota')),
    'Honda': np.array(embeddings.get_text_embedding('Honda')),
    'Ford': np.array(embeddings.get_text_embedding('Ford')),
    'Chevrolet': np.array(embeddings.get_text_embedding('Chevrolet')),
    'Nissan': np.array(embeddings.get_text_embedding('Nissan')),
    'Hyundai': np.array(embeddings.get_text_embedding('Hyundai')),
    'Audi': np.array(embeddings.get_text_embedding('Audi')),
    'Porsche': np.array(embeddings.get_text_embedding('Porsche')),
    'Tesla': np.array(embeddings.get_text_embedding('Tesla')),
    'Apple': np.array(embeddings.get_text_embedding('Apple')),
    'Samsung': np.array(embeddings.get_text_embedding('Samsung')),
    'Google': np.array(embeddings.get_text_embedding('Google')),
    'Facebook': np.array(embeddings.get_text_embedding('Facebook')),
    'Microsoft': np.array(embeddings.get_text_embedding('Microsoft')),
    'Intel': np.array(embeddings.get_text_embedding('Intel')),
    'Nvidia': np.array(embeddings.get_text_embedding('Nvidia')),
    'IBM': np.array(embeddings.get_text_embedding('IBM')),
    'Oracle': np.array(embeddings.get_text_embedding('Oracle')),
    'Amazon': np.array(embeddings.get_text_embedding('Amazon')),
    'Cisco': np.array(embeddings.get_text_embedding('Cisco')),
    'Adobe': np.array(embeddings.get_text_embedding('Adobe'))
}
# Lista de palabras
words = list(embeddings_dic.keys())

# Matriz de embeddings
embedding_matrix = np.array([embeddings_dic[word] for word in words])

# Aplicar PCA para reducir a 2 dimensiones
pca = PCA(n_components=2)
reduced_embeddings = pca.fit_transform(embedding_matrix)

# Graficar
plt.figure(figsize=(10, 8))
for i, word in enumerate(words):
    plt.scatter(reduced_embeddings[i, 0], reduced_embeddings[i, 1])
    plt.text(reduced_embeddings[i, 0] + 0.02, reduced_embeddings[i, 1] + 0.02, word, fontsize=12)

plt.title("Embeddings visualizados en 2D usando PCA")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.grid(True)
plt.show()
```
